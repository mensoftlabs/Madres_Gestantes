```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.integrate import simpson
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from scipy.io import wavfile

def load_and_preprocess(signal_data):
    """
    Si `signal_data` es una cadena, asume que es una ruta de archivo.
    Si `signal_data` es un dict, usa los datos ya procesados.
    """
    if isinstance(signal_data, dict):  # Señal procesada
        return signal_data["sample_rate"], signal_data["time"], signal_data["normalized_data"]
    else:  # Ruta de archivo
        sample_rate, data = wavfile.read(signal_data)
        if len(data.shape) > 1:
            data = np.mean(data, axis=1)
        normalized_data = data / np.max(np.abs(data))
        time = np.arange(0, len(data)) / sample_rate
        return sample_rate, time, normalized_data

def define_intervals(manual_intervals, time):
    intervals = []
    for start_time, end_time in manual_intervals:
        start_idx = np.searchsorted(time, start_time)
        end_idx = np.searchsorted(time, end_time)
        intervals.append((start_idx, end_idx))
    return intervals

def calculate_group_energy(normalized_data, interval, sample_rate):
    start_idx, end_idx = interval
    segment = normalized_data[start_idx:end_idx]  # Include the entire group
    energy = simpson(np.abs(segment)**2, dx=1/sample_rate)  # Use absolute values for integration
    return energy

def create_dataframe(distances, energies_list):
    df = pd.DataFrame({"Distance": distances})
    for i, energies in enumerate(energies_list):
        df[f"Amplitude Test {i+1}"] = energies
    return df

def plot_signal_with_intervals(time, normalized_data, intervals, title):
    plt.figure(figsize=(12, 6))
    plt.plot(time, normalized_data, label="Normalized Signal")
    for start_idx, end_idx in intervals:
        plt.axvspan(time[start_idx], time[end_idx], color='green', alpha=0.3, label="Group Interval" if start_idx == intervals[0][0] else "")
    plt.title(title)
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.legend()
    plt.show()

def plot_distance_vs_energy(df):
    plt.figure(figsize=(10, 6))
    for column in df.columns[1:]:
        plt.scatter(df['Distance'], df[column], label=column)
    plt.title("Distance vs Energy for All Tests")
    plt.xlabel("Distance (cm)")
    plt.ylabel("Energy (Amplitude^2 Integral)")
    plt.grid()
    plt.legend()
    plt.show()

from sklearn.metrics import r2_score
from scipy.optimize import curve_fit

# Función para entrenar el modelo de red neuronal
def train_neural_network(df):
    """
    Entrena una red neuronal sobre los datos del DataFrame.
    """
    X = df[['Distance']].values
    y = df.iloc[:, 1:].mean(axis=1).values.reshape(-1, 1)  # Use the mean of all tests as target

    # Escalar datos
    scaler_X = MinMaxScaler()
    scaler_y = MinMaxScaler()
    X_scaled = scaler_X.fit_transform(X)
    y_scaled = scaler_y.fit_transform(y)

    # Dividir en entrenamiento y prueba
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

    # Crear y entrenar la red neuronal
    model = Sequential([
        Dense(64, activation='relu', input_dim=1),
        Dense(32, activation='relu'),
        Dense(1, activation='linear')
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    model.fit(X_train, y_train, epochs=500, batch_size=8, verbose=0)

    # Predicciones de la red neuronal en el conjunto completo
    all_distances_scaled = scaler_X.transform(X)
    predicted_energies_scaled = model.predict(all_distances_scaled)
    predicted_energies = scaler_y.inverse_transform(predicted_energies_scaled).flatten()

    # Calcular R^2
    r2_nn = r2_score(y, predicted_energies)

    return model, scaler_X, scaler_y, r2_nn, predicted_energies

# Función para ajustar el modelo exponencial
def fit_exponential_model(df):
    """
    Ajusta un modelo exponencial a los datos del DataFrame.
    """
    distances = df['Distance'].values
    energies = df.iloc[:, 1:].mean(axis=1).values  # Promedio de las energías

    # Ajustar el modelo exponencial
    def exponential_decay(d, E0, alpha):
        return E0 * np.exp(-alpha * d)

    popt, _ = curve_fit(exponential_decay, distances, energies, p0=(energies[0], 0.01))
    E0, alpha = popt

    # Predicciones del modelo exponencial
    predicted_energies = exponential_decay(distances, E0, alpha)

    # Calcular R^2
    r2_exponential = r2_score(energies, predicted_energies)

    return E0, alpha, r2_exponential, predicted_energies

# Graficar los datos experimentales y los ajustes
def plot_combined_models(df, predicted_nn, predicted_exp_real, E0_real, alpha_real, r2_nn, r2_exponential):
    """
    Grafica los datos experimentales junto con los ajustes de la red neuronal y el modelo exponencial.
    """
    plt.figure(figsize=(12, 8))

    # Datos experimentales en negro
    for column in df.columns[1:]:
        plt.scatter(df['Distance'], df[column], label=f"Datos: {column}", color="black", edgecolor="black", s=50, alpha=0.6)

    # Ajuste de la Red Neuronal en verde
    dense_distances = np.linspace(df['Distance'].min(), df['Distance'].max(), 100)
    plt.plot(dense_distances, predicted_nn, label=f"Ajuste Red Neuronal (R² = {r2_nn:.3f})", color="green", linewidth=2)

    # Ajuste exponencial en rojo
    plt.plot(dense_distances, predicted_exp_real, label=f"Ajuste Exponencial\n$E(d) = {E0_real:.3f} \\cdot e^{{-{alpha_real:.3f} \\cdot d}}$\nR² = {r2_exponential:.3f}", color="red", linewidth=2)

    # Configuración de la gráfica
    plt.title("Datos y Ajustes: Red Neuronal y Decaimiento Exponencial", fontsize=14, weight="bold")
    plt.xlabel("Distancia (cm)", fontsize=12)
    plt.ylabel("Energía (Amplitude^2 Integral)", fontsize=12)
    plt.legend(fontsize=11)
    plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
    plt.tight_layout()
    plt.show()


# Main Execution
file_paths = [file_path_1_plano, file_path_2_plano, file_path_3_plano]  # Replace with your uploaded files
manual_intervals_list = [
    [(0, 15), (15,25), (25,37), (37,48), (48,59), (59,68), (68,78), (78,91)],  # Intervals for audio1.wav
    [(0, 15), (15,25), (25,37), (37,48), (48,59), (59,68), (68,78), (78,91)],  # Intervals for audio2.wav
    [(0, 15), (15,25), (25,37), (37,48), (48,59), (59,68), (68,78), (78,91)]   # Intervals for audio3.wav
]  # Adjust these intervals manually as needed

distances = [120, 150, 180, 210, 240, 270, 300, 330]  # Distances in cm
all_energies = []

for i, file_path in enumerate(file_paths):
    sample_rate, time, normalized_data = load_and_preprocess(file_path)
    manual_intervals = manual_intervals_list[i]
    intervals = define_intervals(manual_intervals, time)
    energies = [calculate_group_energy(normalized_data, interval, sample_rate) for interval in intervals]
    all_energies.append(energies)

    # Plot signal with intervals for each test
    plot_signal_with_intervals(time, normalized_data, intervals, title=f"Test {i+1}: Group Intervals")

# Create DataFrame with energies from all tests
df = create_dataframe(distances, all_energies)
print("Generated DataFrame:")
print(df)

# Plot distance vs energy for all tests
plot_distance_vs_energy(df)

# Main Execution
# Entrenar la red neuronal
model, scaler_X, scaler_y, r2_nn, predicted_energies_nn = train_neural_network(df)

# Ajustar el modelo exponencial
E0_real, alpha_real, r2_exponential, predicted_energies_exp = fit_exponential_model(df)

# Generar predicciones para la red neuronal y el modelo exponencial
dense_distances = np.linspace(df['Distance'].min(), df['Distance'].max(), 100).reshape(-1, 1)
dense_distances_scaled = scaler_X.transform(dense_distances)
predicted_nn_scaled = model.predict(dense_distances_scaled)
predicted_nn = scaler_y.inverse_transform(predicted_nn_scaled).flatten()
predicted_exp_real = exponential_decay(dense_distances.flatten(), E0_real, alpha_real)

# Mostrar resultados
print(f"R^2 de la Red Neuronal: {r2_nn:.4f}")
print(f"Parámetros del modelo exponencial: E0 = {E0_real:.4f}, alpha = {alpha_real:.4f}, R^2 = {r2_exponential:.4f}")

# Graficar los resultados
plot_combined_models(df, predicted_nn, predicted_exp_real, E0_real, alpha_real, r2_nn, r2_exponential)
```

```python
import numpy as np
import pandas as pd
from scipy.signal import find_peaks
from scipy.ndimage import gaussian_filter1d
from scipy.io import wavfile
from scipy.integrate import simpson, quad
from numpy import trapz
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt

def load_and_preprocess(signal_data):
    """
    Si `signal_data` es una cadena, asume que es una ruta de archivo.
    Si `signal_data` es un dict, usa los datos ya procesados.
    """
    if isinstance(signal_data, dict):  # Señal procesada
        return signal_data["sample_rate"], signal_data["time"], signal_data["normalized_data"]
    else:  # Ruta de archivo
        sample_rate, data = wavfile.read(signal_data)
        if len(data.shape) > 1:
            data = np.mean(data, axis=1)
        normalized_data = data / np.max(data)
        time = np.arange(0, len(data)) / sample_rate
        return sample_rate, time, normalized_data

# Filtrar ruido en la señal
def filter_signal(normalized_data, sigma=3):
    return gaussian_filter1d(normalized_data, sigma=sigma)

# Identificar todos los picos en la señal
def identify_all_peaks(normalized_data, time, height_threshold=0.05):
    peaks, properties = find_peaks(normalized_data, height=height_threshold)
    peak_amplitudes = properties['peak_heights']
    peak_times = time[peaks]
    return [(peak_idx, peak_amplitude, peak_time) for peak_idx, peak_amplitude, peak_time in zip(peaks, peak_amplitudes, peak_times)]

# Dividir los picos en grupos basados en los silencios
def group_peaks_by_silence(peaks, time, min_silence_duration=2):
    groups = []
    current_group = []
    last_peak_time = None

    for peak in peaks:
        _, _, peak_time = peak
        if last_peak_time is None or (peak_time - last_peak_time) > min_silence_duration:
            if current_group:
                groups.append(current_group)
            current_group = [peak]
        else:
            current_group.append(peak)
        last_peak_time = peak_time

    if current_group:
        groups.append(current_group)

    return groups

# Seleccionar el pico más significativo de cada grupo
def select_significant_peaks(groups):
    return [max(group, key=lambda x: x[1]) for group in groups]

# Calcular las energías usando múltiples métodos
def calculate_energy_methods(time_segment, dominant_vibration, sample_rate):
    results = {}
    results["Simpson"] = simpson(dominant_vibration**2, dx=1/sample_rate)
    results["Trapezoidal"] = trapz(dominant_vibration**2, dx=1/sample_rate)
    results["Gaussian Quadrature"], _ = quad(lambda t: np.interp(t, time_segment, dominant_vibration**2),
                                              time_segment[0], time_segment[-1])
    samples = np.random.uniform(time_segment[0], time_segment[-1], 10000)
    results["Monte Carlo"] = np.mean(np.interp(samples, time_segment, dominant_vibration**2)) * (time_segment[-1] - time_segment[0])
    spline = CubicSpline(time_segment, dominant_vibration**2)
    results["Cubic Splines"] = spline.integrate(time_segment[0], time_segment[-1])
    return results

# Crear DataFrame final para cada archivo
def create_dataframe_from_significant_peaks(significant_peaks, time, normalized_data, sample_rate):
    rows = []
    for i, (peak_idx, peak_amplitude, peak_time) in enumerate(significant_peaks, start=1):
        start_idx = max(0, peak_idx - int(0.5 * sample_rate))
        end_idx = min(len(normalized_data), peak_idx + int(0.5 * sample_rate))
        time_segment = time[start_idx:end_idx]
        vibration_segment = normalized_data[start_idx:end_idx]
        energy_results = calculate_energy_methods(time_segment, vibration_segment, sample_rate)
        rows.append({
            "Vibration Number": i,
            "Time (s)": peak_time,
            "Simpson": energy_results["Simpson"],
            "Trapezoidal": energy_results["Trapezoidal"],
            "Gaussian Quadrature": energy_results["Gaussian Quadrature"],
            "Monte Carlo": energy_results["Monte Carlo"],
            "Cubic Splines": energy_results["Cubic Splines"]
        })
    return pd.DataFrame(rows)

# Agregar columna de distancia al DataFrame
def add_distance_column(df, distances):
    """
    Agrega una columna de distancia al DataFrame.

    Parameters:
    - df: DataFrame al que se le agregará la columna.
    - distances: Lista de distancias que se asignarán.

    Returns:
    - df: DataFrame con la columna de distancia añadida.
    """
    if len(df) > len(distances):
        raise ValueError("El número de vibraciones excede el número de distancias proporcionadas.")
    df.insert(1, "Distancia", distances[:len(df)])  # Insertar la columna después de "Vibration Number"
    return df

# Main Execution for Multiple Signals
file_paths = [file_path_1_plano, file_path_2_plano, file_path_3_plano]  # Reemplaza con los nombres reales de tus archivos
height_threshold = 0.05  # Umbral mínimo para detectar picos
min_silence_duration = 2  # Duración mínima de silencio entre grupos (en segundos)
distances = [120, 150, 180, 210, 240, 270, 300, 330]  # Lista de distancias

dataframes = {}
for i, file_path in enumerate(file_paths, start=1):
    print(f"Processing file: {file_path}")
    sample_rate, time, normalized_data = load_and_preprocess(file_path)
    smoothed_signal = filter_signal(normalized_data)
    all_peaks = identify_all_peaks(smoothed_signal, time, height_threshold)
    groups = group_peaks_by_silence(all_peaks, time, min_silence_duration)
    significant_peaks = select_significant_peaks(groups)
    df = create_dataframe_from_significant_peaks(significant_peaks, time, smoothed_signal, sample_rate)

    # Agregar columna de distancia
    df = add_distance_column(df, distances)
    dataframes[f"Audio {i}"] = df

    # Graficar señal con picos significativos
    plt.figure(figsize=(12, 6))
    plt.plot(time, np.abs(smoothed_signal), label="Normalized Signal (Absolute)")
    for group in groups:
        group_times = [p[2] for p in group]
        plt.axvspan(group_times[0], group_times[-1], color='green', alpha=0.3, label="Vibration Group" if group == groups[0] else "")
    plt.scatter([p[2] for p in significant_peaks], [p[1] for p in significant_peaks], color='red', label="Significant Peaks")
    plt.title(f"Significant Peaks and Vibration Groups for {file_path}")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.legend()
    plt.show()

def add_id_column(df, id_value):
    """
    Agrega una columna de ID al DataFrame.

    Parameters:
    - df: DataFrame al que se le agregará la columna.
    - id_value: Valor del ID que se asignará a todas las filas del DataFrame.

    Returns:
    - df: DataFrame con la columna ID añadida.
    """
    df.insert(0, "ID", id_value)  # Insertar la columna ID al inicio
    return df

# Unir los DataFrames y agregar la columna ID
def combine_dataframes_with_id(dataframes):
    """
    Une múltiples DataFrames en uno solo y agrega una columna de ID para cada DataFrame.

    Parameters:
    - dataframes: Diccionario con nombres como claves y DataFrames como valores.

    Returns:
    - combined_df: DataFrame combinado con una columna ID.
    """
    combined_df = pd.DataFrame()  # DataFrame vacío para la unión
    for i, (key, df) in enumerate(dataframes.items(), start=1):
        df_with_id = add_id_column(df, i)  # Agregar la columna ID
        combined_df = pd.concat([combined_df, df_with_id], ignore_index=True)  # Concatenar
    return combined_df

# Main Execution for Combining DataFrames
combined_df = combine_dataframes_with_id(dataframes)

#----------------------REPRESENTACION----------------------------------------------------------------------------------------------------

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Función exponencial para el ajuste
def exponential_decay(d, E0, alpha):
    """
    Modelo de decaimiento exponencial: E(d) = E0 * exp(-alpha * d)
    """
    return E0 * np.exp(-alpha * d)

# Ajustar el modelo para cada método de integración
def fit_exponential_decay(df, method_column):
    """
    Ajusta el modelo exponencial a los datos para un método de integración específico
    y calcula el coeficiente R^2.
    """
    distances = df["Distancia"].values
    energies = df[method_column].values

    # Ajustar el modelo usando curve_fit
    popt, _ = curve_fit(exponential_decay, distances, energies, p0=(energies[0], 0.01))
    E0, alpha = popt

    # Calcular las energías predichas
    predicted_energies = exponential_decay(distances, E0, alpha)

    # Calcular R^2
    residuals = energies - predicted_energies
    ss_res = np.sum(residuals**2)  # Suma de los residuales al cuadrado
    ss_tot = np.sum((energies - np.mean(energies))**2)  # Suma total al cuadrado
    r2 = 1 - (ss_res / ss_tot)

    return E0, alpha, r2

# Visualizar el ajuste del modelo
def plot_exponential_fit(df, method_column, E0, alpha, method_name):
    """
    Grafica los datos experimentales y la curva ajustada.
    """
    distances = df["Distancia"].values
    energies = df[method_column].values

    # Generar puntos densos para una línea de ajuste suave
    dense_distances = np.linspace(min(distances), max(distances), 100)
    fitted_energies = exponential_decay(dense_distances, E0, alpha)

    plt.figure(figsize=(10, 6))
    plt.scatter(distances, energies, label="Datos Experimentales", color="blue", edgecolor="black", s=70, zorder=3)
    plt.plot(dense_distances, fitted_energies, label=f"Modelo Ajustado\n$E(d) = {E0:.3f} \\cdot e^{{-{alpha:.3f} \\cdot d}}$", color="red", linewidth=2, zorder=2)
    plt.title(f"Ajuste de Decaimiento Exponencial ({method_name})", fontsize=14, weight="bold")
    plt.xlabel("Distancia (cm)", fontsize=12)
    plt.ylabel("Energía", fontsize=12)
    plt.legend(fontsize=11)
    plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
    plt.tight_layout()
    plt.show()

# Main Execution
method_columns = ["Simpson", "Trapezoidal", "Gaussian Quadrature", "Monte Carlo", "Cubic Splines"]
coefficients = []

for method in method_columns:
    print(f"\nCalculando para el método: {method}")
    E0, alpha, r2 = fit_exponential_decay(combined_df, method)
    coefficients.append({"Método": method, "E0": E0, "Alpha (Coeficiente)": alpha, "R2": r2})
    print(f"E0 (Energía Inicial): {E0:.4f}, Alpha (Coeficiente del Medio): {alpha:.4f}, R2: {r2:.4f}")

    # Graficar el ajuste
    plot_exponential_fit(combined_df, method, E0, alpha, method)

# Crear un DataFrame con los resultados
results_df = pd.DataFrame(coefficients)
print("\nResultados:")
print(results_df)
```

```python
import numpy as np
import pandas as pd
from scipy.signal import find_peaks
from scipy.ndimage import gaussian_filter1d
from scipy.io import wavfile
from scipy.integrate import simpson, quad
from numpy import trapz
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt

def load_and_preprocess(signal_data):
    """
    Si `signal_data` es una cadena, asume que es una ruta de archivo.
    Si `signal_data` es un dict, usa los datos ya procesados.
    """
    if isinstance(signal_data, dict):  # Señal procesada
        return signal_data["sample_rate"], signal_data["time"], signal_data["normalized_data"]
    else:  # Ruta de archivo
        sample_rate, data = wavfile.read(signal_data)
        if len(data.shape) > 1:
            data = np.mean(data, axis=1)
        normalized_data = data / np.max(np.abs(data))
        time = np.arange(0, len(data)) / sample_rate
        return sample_rate, time, normalized_data

# Filtrar ruido en la señal
def filter_signal(normalized_data, sigma=3):
    return gaussian_filter1d(normalized_data, sigma=sigma)

# Identificar todos los picos en la señal
def identify_all_peaks(normalized_data, time, height_threshold=0.05):
    peaks, properties = find_peaks(np.abs(normalized_data), height=height_threshold)
    peak_amplitudes = properties['peak_heights']
    peak_times = time[peaks]
    return [(peak_idx, peak_amplitude, peak_time) for peak_idx, peak_amplitude, peak_time in zip(peaks, peak_amplitudes, peak_times)]

# Dividir los picos en grupos basados en los silencios
def group_peaks_by_silence(peaks, time, min_silence_duration=2):
    groups = []
    current_group = []
    last_peak_time = None

    for peak in peaks:
        _, _, peak_time = peak
        if last_peak_time is None or (peak_time - last_peak_time) > min_silence_duration:
            if current_group:
                groups.append(current_group)
            current_group = [peak]
        else:
            current_group.append(peak)
        last_peak_time = peak_time

    if current_group:
        groups.append(current_group)

    return groups

# Seleccionar el pico más significativo de cada grupo
def select_significant_peaks(groups):
    return [max(group, key=lambda x: x[1]) for group in groups]

# Calcular las energías usando múltiples métodos
def calculate_energy_methods(time_segment, dominant_vibration, sample_rate):
    results = {}
    results["Simpson"] = simpson(dominant_vibration**2, dx=1/sample_rate)
    results["Trapezoidal"] = trapz(dominant_vibration**2, dx=1/sample_rate)
    results["Gaussian Quadrature"], _ = quad(lambda t: np.interp(t, time_segment, dominant_vibration**2),
                                              time_segment[0], time_segment[-1])
    samples = np.random.uniform(time_segment[0], time_segment[-1], 10000)
    results["Monte Carlo"] = np.mean(np.interp(samples, time_segment, dominant_vibration**2)) * (time_segment[-1] - time_segment[0])
    spline = CubicSpline(time_segment, dominant_vibration**2)
    results["Cubic Splines"] = spline.integrate(time_segment[0], time_segment[-1])
    return results

# Crear DataFrame final para cada archivo
def create_dataframe_from_significant_peaks(significant_peaks, time, normalized_data, sample_rate):
    rows = []
    for i, (peak_idx, peak_amplitude, peak_time) in enumerate(significant_peaks, start=1):
        start_idx = max(0, peak_idx - int(0.5 * sample_rate))
        end_idx = min(len(normalized_data), peak_idx + int(0.5 * sample_rate))
        time_segment = time[start_idx:end_idx]
        vibration_segment = normalized_data[start_idx:end_idx]
        energy_results = calculate_energy_methods(time_segment, vibration_segment, sample_rate)
        rows.append({
            "Vibration Number": i,
            "Time (s)": peak_time,
            "Simpson": energy_results["Simpson"],
            "Trapezoidal": energy_results["Trapezoidal"],
            "Gaussian Quadrature": energy_results["Gaussian Quadrature"],
            "Monte Carlo": energy_results["Monte Carlo"],
            "Cubic Splines": energy_results["Cubic Splines"]
        })
    return pd.DataFrame(rows)

# Agregar columna de distancia al DataFrame
def add_distance_column(df, distances):
    """
    Agrega una columna de distancia al DataFrame.

    Parameters:
    - df: DataFrame al que se le agregará la columna.
    - distances: Lista de distancias que se asignarán.

    Returns:
    - df: DataFrame con la columna de distancia añadida.
    """
    if len(df) > len(distances):
        raise ValueError("El número de vibraciones excede el número de distancias proporcionadas.")
    df.insert(1, "Distancia", distances[:len(df)])  # Insertar la columna después de "Vibration Number"
    return df

# Main Execution for Multiple Signals
file_paths = [file_path_1_circular, file_path_2_circular, file_path_3_circular]  # Reemplaza con los nombres reales de tus archivos
height_threshold = 0.05  # Umbral mínimo para detectar picos
min_silence_duration = 2  # Duración mínima de silencio entre grupos (en segundos)
distances = [120, 150, 180, 210, 240, 270, 300, 330]  # Lista de distancias

dataframes = {}
for i, file_path in enumerate(file_paths, start=1):
    print(f"Processing file: {file_path}")
    sample_rate, time, normalized_data = load_and_preprocess(file_path)
    smoothed_signal = filter_signal(normalized_data)
    all_peaks = identify_all_peaks(smoothed_signal, time, height_threshold)
    groups = group_peaks_by_silence(all_peaks, time, min_silence_duration)
    significant_peaks = select_significant_peaks(groups)
    df = create_dataframe_from_significant_peaks(significant_peaks, time, smoothed_signal, sample_rate)

    # Agregar columna de distancia
    df = add_distance_column(df, distances)
    dataframes[f"Audio {i}"] = df

    # Graficar señal con picos significativos
    plt.figure(figsize=(12, 6))
    plt.plot(time, np.abs(smoothed_signal), label="Normalized Signal (Absolute)")
    for group in groups:
        group_times = [p[2] for p in group]
        plt.axvspan(group_times[0], group_times[-1], color='green', alpha=0.3, label="Vibration Group" if group == groups[0] else "")
    plt.scatter([p[2] for p in significant_peaks], [p[1] for p in significant_peaks], color='red', label="Significant Peaks")
    plt.title(f"Significant Peaks and Vibration Groups for {file_path}")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.legend()
    plt.show()


def add_id_column(df, id_value):
    """
    Agrega una columna de ID al DataFrame.

    Parameters:
    - df: DataFrame al que se le agregará la columna.
    - id_value: Valor del ID que se asignará a todas las filas del DataFrame.

    Returns:
    - df: DataFrame con la columna ID añadida.
    """
    df.insert(0, "ID", id_value)  # Insertar la columna ID al inicio
    return df

# Unir los DataFrames y agregar la columna ID
def combine_dataframes_with_id(dataframes):
    """
    Une múltiples DataFrames en uno solo y agrega una columna de ID para cada DataFrame.

    Parameters:
    - dataframes: Diccionario con nombres como claves y DataFrames como valores.

    Returns:
    - combined_df: DataFrame combinado con una columna ID.
    """
    combined_df = pd.DataFrame()  # DataFrame vacío para la unión
    for i, (key, df) in enumerate(dataframes.items(), start=1):
        df_with_id = add_id_column(df, i)  # Agregar la columna ID
        combined_df = pd.concat([combined_df, df_with_id], ignore_index=True)  # Concatenar
    return combined_df

# Main Execution for Combining DataFrames
combined_df = combine_dataframes_with_id(dataframes)

#----------------------REPRESENTACION----------------------------------------------------------------------------------------------------

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Función exponencial para el ajuste
def exponential_decay(d, E0, alpha):
    """
    Modelo de decaimiento exponencial: E(d) = E0 * exp(-alpha * d)
    """
    return E0 * np.exp(-alpha * d)

# Ajustar el modelo para cada método de integración
def fit_exponential_decay(df, method_column):
    """
    Ajusta el modelo exponencial a los datos para un método de integración específico.
    """
    distances = df["Distancia"].values
    energies = df[method_column].values

    # Ajustar el modelo usando curve_fit
    popt, _ = curve_fit(exponential_decay, distances, energies, p0=(energies[0], 0.01))
    E0, alpha = popt
    return E0, alpha

# Visualizar el ajuste del modelo
def plot_exponential_fit(df, method_column, E0, alpha, method_name):
    """
    Grafica los datos experimentales y la curva ajustada.
    """
    distances = df["Distancia"].values
    energies = df[method_column].values

    # Generar puntos densos para una línea de ajuste suave
    dense_distances = np.linspace(min(distances), max(distances), 100)
    fitted_energies = exponential_decay(dense_distances, E0, alpha)

    plt.figure(figsize=(10, 6))
    plt.scatter(distances, energies, label="Datos Experimentales", color="blue", edgecolor="black", s=70, zorder=3)
    plt.plot(dense_distances, fitted_energies, label=f"Modelo Ajustado\n$E(d) = {E0:.3f} \\cdot e^{{-{alpha:.3f} \\cdot d}}$", color="red", linewidth=2, zorder=2)
    plt.title(f"Ajuste de Decaimiento Exponencial ({method_name})", fontsize=14, weight="bold")
    plt.xlabel("Distancia (cm)", fontsize=12)
    plt.ylabel("Energía", fontsize=12)
    plt.legend(fontsize=11)
    plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
    plt.tight_layout()
    plt.show()

# Main Execution
method_columns = ["Simpson", "Trapezoidal", "Gaussian Quadrature", "Monte Carlo", "Cubic Splines"]
coefficients = []

for method in method_columns:
    print(f"\nCalculando para el método: {method}")
    E0, alpha = fit_exponential_decay(combined_df, method)
    coefficients.append({"Método": method, "E0": E0, "Alpha (Coeficiente)": alpha})
    print(f"E0 (Energía Inicial): {E0:.4f}, Alpha (Coeficiente del Medio): {alpha:.4f}")

    # Graficar el ajuste
    plot_exponential_fit(combined_df, method, E0, alpha, method)

# Crear un DataFrame con los resultados
results_df = pd.DataFrame(coefficients)
print("\nResultados:")
print(results_df)
```
